```
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus
from xml.etree import ElementTree
import pandas as pd

skey = 'crc%2FlWi6vrE%2BSqINyCZza2oF2mn0YET3A1EV5Rl3OION0X71d0C8BMhrhL0PkJMESoz61l246P4M8k%2BJyzOotw%3D%3D'

url = 'http://openapi.tago.go.kr/openapi/service/SubwayInfoService/getKwrdFndSubwaySttnList'
params = '?' + 'ServiceKey=' + skey + '&' + \
          urlencode({quote_plus('subwayStationName') : '서울'})
print(url + params)

request = Request(url + params)
request.get_method = lambda: 'GET'
response_body = urlopen(request).read()
#print(response_body)

root = ElementTree.fromstring(response_body)

df = pd.DataFrame()
for item in root.iter('item'):
  item_dict = {}
  item_dict['subwayRouteName'] = item.find('subwayRouteName').text
  item_dict['subwayStationId'] = item.find('subwayStationId').text
  item_dict['subwayStationName'] = item.find('subwayStationName').text
  df = df.append(item_dict, ignore_index=True)

df
```
